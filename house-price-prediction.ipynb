## Feature Engineering

Several domain-informed features were created to improve predictive performance:

- **TotalSF**: total square footage of the house
- **HouseAge**: age of the house at the time of sale
- **RemodAge**: time since last remodeling
- **TotalBathrooms**: combined bathroom count with weighting for half baths
- **TotalAmenities**: count of major binary amenities such as garage, pool, basement, fireplace, and second floor
- **Qual_x_SF**: interaction between overall quality and living area

These features capture structural size, age-related depreciation, and quality-size interactions more effectively than raw variables alone.

## Missing Value Strategy

Missing values were handled according to feature semantics:

- Categorical features describing absence of a structure were filled with `"None"`
- Numerical features related to absent structures were filled with `0`
- Some features were imputed using group-based statistics or mode/median values

- Mode filling for categorical variables
- Median filling for numerical variables
- Group-based imputation

## 8. Feature Engineering

We construct new features to improve model performance.

### Numerical combinations
- Total square footage
- Total bathrooms
- Total porch area

### Time features
- HouseAge
- RemodAge

### Binary features
- HasGarage, HasPool, HasBsmt, etc.

### Interaction features
- OverallQual × GrLivArea

### Skewed features processing
- Several area-based features are right-skewed, suggesting the need for log transformation.

```python
all_features['YrSold'] = all_features['YrSold'].astype(int)
all_features['YearBuilt'] = all_features['YearBuilt'].astype(int)
all_features['YearRemodAdd'] = all_features['YearRemodAdd'].astype(int)
def create_features(df):
    df = df.copy()

    # ======================
    # math combination
    # ======================
    df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']

    df['Total_sqr_footage'] = (
        df['BsmtFinSF1'] + df['BsmtFinSF2'] +
        df['1stFlrSF'] + df['2ndFlrSF']
    )

    df['Total_porch_sf'] = (
        df['OpenPorchSF'] + df['3SsnPorch'] +
        df['EnclosedPorch'] + df['ScreenPorch'] +
        df['WoodDeckSF']
    )

    df['Total_Bathrooms'] = (
        df['FullBath'] +
        0.5 * df['HalfBath'] +
        df['BsmtFullBath'] +
        0.5 * df['BsmtHalfBath']
    )

    df['HouseAge'] = df['YrSold'] - df['YearBuilt']
    df['RemodAge'] = df['YrSold'] - df['YearRemodAdd']

    df['Total_Home_Quality'] = df['OverallQual'] + df['OverallCond']

    # ======================
    # binary features
    # ======================
    df['HasPool'] = (df['PoolArea'] > 0).astype(int)
    df['HasGarage'] = (df['GarageArea'] > 0).astype(int)
    df['HasBsmt'] = (df['TotalBsmtSF'] > 0).astype(int)
    df['HasFireplace'] = (df['Fireplaces'] > 0).astype(int)
    df['Has2ndFloor'] = (df['2ndFlrSF'] > 0).astype(int)

    df['TotalAmenities'] = (
        df['HasPool'] +
        df['HasGarage'] +
        df['HasFireplace'] +
        df['Has2ndFloor']
    )

    df['Qual_x_SF'] = df['OverallQual'] * df['GrLivArea']

    # ======================
    # skewed feature handling
    # ======================
    numeric_cols = df.select_dtypes(include=['int16', 'int32', 'int64', 'float16', 'float32', 'float64']).columns

    continuous_cols = [col for col in numeric_cols if df[col].nunique() > 20]

    skew_values = df[continuous_cols].apply(lambda x: skew(x.dropna()))

    skew_features = skew_values[skew_values > 0.5].sort_values(ascending=False)
    

    log_features = list(skew_features.index)


    exclude_cols = {'HouseAge', 'RemodAge'}
    log_features = [col for col in log_features if col not in exclude_cols]

    for col in log_features:
        if col + '_log' not in df.columns:
            df[col + '_log'] = np.log1p(df[col])

    return df

all_features = create_features(all_features)
```

## 9. Encoding

- Apply one-hot encoding using `pd.get_dummies`
- Remove duplicate columns

```python
all_features = pd.get_dummies(all_features).reset_index(drop=True)
all_features = all_features.loc[:,~all_features.columns.duplicated()]
all_features = all_features.drop(['Id'], axis=1)
```

```python
X_train = all_features.iloc[:len(train_labels), :]
X_test  = all_features.iloc[len(train_labels):, :]
```

## 10. Model Building

We compare several regression models with complementary inductive biases:

- **Ridge Regression** serves as a strong linear baseline and works well with sparse one-hot encoded features.
- **Random Forest** captures nonlinear relationships through bagging and feature subsampling.
- **Gradient Boosting Regressor** models complex interactions by sequentially correcting residual errors.
- **LightGBM** is a high-performance gradient boosting model that often performs well on structured tabular data.

To improve generalization, these base learners are combined using a stacking ensemble.  
A **Ridge Regression** meta-learner is used as the final estimator because it is simple, regularized, and effective at combining correlated base predictions without overfitting.

```python
# Cross-validation strategy
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Base models
base_models = [
    ('ridge', Ridge(alpha=10.0)),
    ('rf', RandomForestRegressor(
        n_estimators=300,
        random_state=42,
        n_jobs=-1
    )),
    ('gbr', GradientBoostingRegressor(
        n_estimators=300,
        learning_rate=0.05,
        random_state=42
    )),
    ('lgb', LGBMRegressor(
        n_estimators=3000,
        learning_rate=0.01,
        num_leaves=31,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1
    ))
]

# Helper function for CV evaluation
def rmse_cv(model, X, y):
    scores = cross_val_score(
        model,
        X,
        y,
        scoring='neg_root_mean_squared_error',
        cv=cv,
        n_jobs=-1
    )
    return -scores.mean(), scores.std()

# Evaluate base models
results = []
for name, model in base_models:
    mean_rmse, std_rmse = rmse_cv(model, X_train, train_labels)
    results.append([name, mean_rmse, std_rmse])

results_df = pd.DataFrame(results, columns=["Model", "CV_RMSE", "Std"])
results_df.sort_values("CV_RMSE", inplace=True)
results_df
```

## 11. Stacking Ensemble

We combine multiple models using stacking:

### Base Models
- Ridge
- Random Forest
- Gradient Boosting
- LightGBM

### Final Model
- Ridge Regression

Cross-validation is used to improve generalization.

```python
stack_model = StackingRegressor(
    estimators=base_models,
    final_estimator=Ridge(),
    cv=5,
    n_jobs=-1
)
```

```python
model_names = []
cv_means = []
cv_stds = []

for name, model in base_models:
    mean_rmse, std_rmse = rmse_cv(model, X_train, train_labels)
    model_names.append(name)
    cv_means.append(mean_rmse)
    cv_stds.append(std_rmse)

comparison_df = pd.DataFrame({
    "Model": model_names,
    "CV_RMSE": cv_means,
    "Std": cv_stds
}).sort_values("CV_RMSE")

comparison_df
```

Cross-validation shows the relative strength of each base learner.  
The stacking ensemble is then used to combine their complementary strengths and improve robustness on unseen data.

## 12. Prediction and Submission

- Generate predictions on test data
- Apply inverse transformation using `expm1`
- Export results for Kaggle submission

```python
stack_model.fit(X_train, train_labels)

final_pred = stack_model.predict(X_test)

final_pred = np.expm1(final_pred)

submission = pd.DataFrame({ "Id": test_df["Id"], "SalePrice": final_pred })

submission.to_csv("submission.csv", index=False)
```
